The .next() method returns the current row and moves to the next row.

The first time you run next() it returns the header and the next time you run it returns the first record and so on.

````python
try:
  header = []
  header = next(csvreader)
  print(header)
except Exception as e:
  log.error("Error at printing header!!!")
  log.exception(e)
````

**4. What method takes a list argument and writes it to a CSV file?**

writelines() : Iterate through each list and convert the list elements to a string and write to the csv file.

````python
###This is simple example of write() method
header = ['Name', 'M1 Score', 'M2 Score']
data = [['Alex', 62, 80], ['Brad', 45, 56], ['Joey', 85, 98]]
filename = 'Student_scores.csv'
with open(filename, 'w') as file:
    for head in header:
        file.write(str(head)+', ')
    file.write('\n')
    for row in data:
        for x in row:
            file.write(str(x)+', ')
        file.write('\n')
````

````python

# Writing a file using writelines()
f = open(filename, 'a+')
f.seek(0)
lines = f.readlines()   #Read the data from the above file created by write()
print(lines)  
#writelines()
f.writelines(lines)    #Now replace the data using writelines()
f.close()
````

````python
# Reading a file generated by writelines()
f = open(filename, 'r')
lines = f.readlines()
print(lines)

f.close()
````

**5. What do the keyword arguments delimiter and line terminator do?**

delimiter : separate the cell based on given delimeter, default is comma ","


lineterminator: By default it is new line "\n", but this can be changed by giving the required parameter.



````python
import csv

csvFile = open('example1.tsv', 'w', newline='')
#delimiter is tab "\t" and lineterminator is double-spaced "\n\n"
csvWriter = csv.writer(csvFile, delimiter='\t', lineterminator='\n\n')
csvWriter.writerow(['abc', 'def', 'ghi'])
csvWriter.writerow(['eggs', 'bacon', 'ham'])
csvWriter.writerow(['spam', 'spam', 'spam', 'spam', 'spam', 'spam'])
csvWriter.writerow(['spam', 'spam', 'spam', 1,2,3,6,5,4])
csvFile.close()
````

**6. What function takes a string of JSON data and returns a Python data structure?**

After we import the json module, we can call loads() and pass it a string of JSON data. (Note that JSON strings always use double quotes). It will return that data as a Python dictionary. 

````python
import json
strJSONData= '{"name":"John", "age":30, "car":null}'

out = json.loads(strJSONData)  #loads --> load string

print(type(out))
````
